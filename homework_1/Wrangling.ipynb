```python
# TODO: get random sample of flights from SQL
flights_s_df = pd.read_sql_query('SELECT * FROM flights ORDER BY RANDOM() LIMIT 1000', con=engine)
```

```python
# TODO: create nested map of Jaccard distance
```

first I convered the `ArrDelayMinutes` into float type because it was of type `object`, but it was not categorical.

```python
def str2float(s):
    if type(s) == str:
        if len(s) == 0:
            return(float('nan'))
        else:
            return (float(s))
    else:
        return(s)

flights_s_df['ArrDelayMinutes']=[str2float(x) for x in flights_s_df['ArrDelayMinutes']]
flights_s_df.dtypes
```

then I ran the jaccard similarity score

```python
def jicard_score(list1,list2):
    cleaned_list1=[x for x in list1 if str(x) != 'nan']
    cleaned_list2=[x for x in list2 if str(x) != 'nan']
    set1=set(cleaned_list1)
    set2=set(cleaned_list2)
    intersect = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return(intersect/union)

jicard_dist={}
for col1 in routes_df.columns:
    if routes_df[col1].dtype != np.float64 and \
    routes_df[col1].dtype != np.int64:
        df1 = routes_df.drop_duplicates(col1)
        jicard_dist[col1]={}
        for col2 in flights_s_df.columns:
            if flights_s_df[col2].dtype != np.float64 and \
            flights_s_df[col2].dtype != np.int64:
                df2 = flights_s_df.drop_duplicates(col2)
                jicard_s = jicard_score(df1[col1], df2[col2])
                if jicard_s != 0:
                    jicard_dist[col1][col2]=jicard_s
```

## Step 5.2.2

```python
# TODO: output bad flights
badflights_df = pd.DataFrame(columns=['airline_dest','badflight_count'])

for name, group in flights_s_df.groupby(["airline_iata_icao","dest_airport_iata_icao"]):
    #print(name)
    nCount = len([True for index, row in group.iterrows() if \
           row["ArrDelayMinutes"] >= 30.0 or \
           row["Cancelled"] == 1.0])
    this_incident_df = pd.DataFrame([[name, nCount]], \
                                   columns=['airline_dest','badflight_count'])
    
    badflights_df = pd.concat([badflights_df, this_incident_df], ignore_index=True)
badflights_df= badflights_df.set_index('airline_dest')
badflights_df
```
