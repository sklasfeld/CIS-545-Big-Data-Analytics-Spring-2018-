### Step 6.2.2

```python
# TODO: Create a DataFrame for each airline IATA code 
# and destination city IATA code, the number of flights
# (from the data of Section 6.2.1), whose arrival was 
# delayed by 30 minutes or more, or which were cancelled. 

cancelled_df = pd.read_sql("SELECT airline_iata_icao, \
    dest_airport_iata_icao, COUNT(*) \
    FROM routes JOIN flights \
    ON airline_iata_icao = Carrier AND \
    dest_airport_iata_icao = Dest AND \
    source_airport_iata_icao = Origin \
    WHERE ArrDelayMinutes >= 30.0 OR \
    Cancelled = 1.0 \
    GROUP BY airline_iata_icao, \
    dest_airport_iata_icao \
    ORDER BY airline_iata_icao, Dest LIMIT 60", \
    engine)
cancelled_df
```

### Step 6.2.3

```python
# TODO: Create a DataFrame  based off the one in 
# 6.2.2 but with the HAVING constraint specifying
# that the COUNT must be at least 1000 flights 
# delayed by 30 minutes or more, or cancelled.  
most_delayed_df = pd.read_sql("SELECT airline_iata_icao, \
    dest_airport_iata_icao, COUNT(*) \
    FROM routes JOIN flights \
    ON airline_iata_icao = Carrier AND \
    dest_airport_iata_icao = Dest AND \
    source_airport_iata_icao = Origin \
    WHERE ArrDelayMinutes >= 30.0 OR \
    Cancelled = 1.0 \
    GROUP BY airline_iata_icao, \
    dest_airport_iata_icao \
    HAVING COUNT(*)>=1000 \
    ORDER BY airline_iata_icao, Dest LIMIT 60", \
    engine)
most_delayed_df
```

```python
# TODO: Create a “base case” where you take compute, 
# and store in a SQL table, the earliest arrival for
# each departure time for each pair of cities on the
# same day.  Make sure that all flights in the table 
# are along a valid route by joining with the routes 
# table (like in previous parts of this assignment).  
# Put these these into a table called shortest.
shortest_df = pd.read_sql("SELECT Year, Month, DayofMonth, \
    Dest, Origin, DepTime, MIN(ArrTime) AS MinArrTime \
    FROM routes JOIN flights \
    ON airline_iata_icao = Carrier AND \
    dest_airport_iata_icao = Dest AND \
    source_airport_iata_icao = Origin \
    WHERE Year IS NOT NULL AND \
    Month=9 AND \
    DayofMonth<9 AND \
    ArrTime IS NOT NULL AND DepTime \
    IS NOT NULL AND DepTime != 2400.0 \
    AND DepTime < ArrTime \
    AND (source_airport_iata_icao ='LAX' \
    OR dest_airport_iata_icao = 'DEN') \
    GROUP BY DepTime, Dest, Origin, Year, Month, \
    DayofMonth", \
    engine)

shortest_df.to_sql("shortest", con=engine, if_exists='replace', index=False)

#most_delayed_df
```

```python
# TODO: find a pair of shorter paths that leave 
# at the same time, but arrive earlier to the 
# same destination on the same day. 

pd.read_sql("SELECT s1.Origin, s1.Dest AS Layover, s2.Dest, \
    s1.DepTime, s2.MinArrTime \
    FROM shortest AS s1 JOIN shortest as s2 \
    ON s1.DayofMonth=s2.DayofMonth AND \
    s2.Origin = s1.Dest \
    WHERE s1.MinArrTime + 100.0 <= s2.DepTime AND \
    EXISTS \
        (SELECT s3.Origin FROM shortest as s3 WHERE \
        s3.Year=s1.Year AND \
        s3.Month=s1.Month AND \
        s3.DayofMonth=s1.DayofMonth AND \
        s3.Origin=s1.Origin AND \
        s3.MinArrTime>s2.MinArrTime)", \
    engine)
    
#merged_shortest_df.to_sql("merged_shortest", con=engine, if_exists='replace', index=False)


#merged_shortest_df["ArrTime"]
#most_delayed_df
```
