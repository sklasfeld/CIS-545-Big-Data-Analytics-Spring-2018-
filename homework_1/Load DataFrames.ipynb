```python
# TODO: Load airports.dat.txt into a dataframe called airports_df
airports_df_colnames = ["airport_id", "airport_name", \
                       "city", "country", \
                       "airport_iata", "airport_icao", "latitude", \
                       "longitude", \
                        "altitude", \
                        "timezone", "dst", \
                        "tz"]
airports_df = pd.read_csv("airports.dat.txt", \
                          header=None, \
                          names = airports_df_colnames)
```

```python
# correct for daylight savings by country
country_dst={}
for index, row in airports_df.iterrows():
    if row["country"] in country_dst:
        if country_dst[row["country"]] == "N" or \
        country_dst[row["country"]] == "U":
            country_dst[row["country"]]=row["dst"]
    else:
        country_dst[row["country"]]=row["dst"]
        
for each_country in airports_df.country.unique():
    if country_dst[each_country] != "N" and \
        country_dst[each_country] != "U":
        airports_df.loc[airports_df["country"]==each_country,"dst"]=country_dst[each_country]
```

```python
# fillna

# Replace NaNs with blanks if the column is a string, so we donâ€™t get
# objects in the column
def fillna_col(series):
    if series.dtype is pd.np.dtype(object):
        return series.fillna('')
    else:
        return series
```

```python
# TODO: Define nullify here

# Replace \Ns with NaNs
def nullify(x):
    if x == r"\N":
        return(float('NaN'))  # TODO: replace this!
    else:
        return(x)
```

```python
# TODO: regularize and remove nulls according to Step 2
```

```python
# apply nullify to all of the elements in each of 
# our DataFrames 
routes_df = routes_df.applymap(nullify)
airlines_df = airlines_df.applymap(nullify)
flights_df = flights_df.applymap(nullify)
airports_df = airports_df.applymap(nullify)
```

```python
# get rid of rows in routes_df that have 
# null airline, source, or destination IDs
routes_df.dropna(axis=0, how="any", \
                 subset=["airline_id", \
                         "source_airport_id", \
                         "dest_airport_id"], \
                 inplace=True)
```

```python
# apply fillna_col to all of the columns in 
# each of our DataFrames 
routes_df = routes_df.apply(fillna_col, axis=1)
airlines_df = airlines_df.apply(fillna_col, axis=1)
flights_df = flights_df.apply(fillna_col, axis=1)
airports_df = airports_df.apply(fillna_col, axis=1)
```

```python
# TODO: change column types according to Step 2
routes_df['airline_id'] = routes_df['airline_id'].astype(int)
routes_df['source_airport_id'] = routes_df['source_airport_id'].astype(int)
routes_df['dest_airport_id'] = routes_df['dest_airport_id'].astype(int)
```

```python
# TODO: read and clean raw text file and store in clean_incidents
import re
clean_incidents=[]
year=0
months=["January","February","March", \
        "April","May","June", "July", \
        "August","September","October",\
        "November", "December"]
with open("incidents_raw.txt") as f:
    for line in f:
        if re.match(r"[0-9][0-9][0-9][0-9]\[edit\]",line):
            year=int(line.split("[")[0])
        else:
            if year >= 1997:
                fields=line.split(" ")
                if fields[0] in months:
                    if re.match(r"[0-9]*", fields[1]) and \
                    int(fields[1])<32 and \
                    int(fields[1])>0 and \
                    fields[2] == "â€“":
                            incident = ("%i %s %s , %s" % \
                                       (year, fields[0],
                                       fields[1]," ".join(fields[3:])))
                            clean_incidents.append(incident)
```

```python
# TODO: convert clean_incidents into dataframe incidents_df
from datetime import datetime

incidents_df = pd.DataFrame(columns=['Date', 'Airline', 'FlightNum'])

airline_name_list =list(set(airlines_df["airline_name"]))
for line in clean_incidents:
    fields=line.split(" ")
    dateOfInc = ("%s %s %s" % (fields[0], fields[1], fields[2]))
    dtOfInc = datetime.strptime(dateOfInc, '%Y %B %d')
    inc_airline = float('NaN')
    for aline_name in airline_name_list:
        aline_flight = ("%s Flight" % (aline_name))
        if aline_flight.upper() in line.upper():
            
                if inc_airline != inc_airline :
                    inc_airline = aline_name
                elif len(aline_name)>len(inc_airline):
                    inc_airline = aline_name
    inc_flight=float('NaN')
    if "Flight" in line:
        flight_idx = fields.index("Flight")
        if inc_airline != inc_airline:
            inc_airline=" ".join(fields[4:flight_idx])
        if re.match(r"[0-9]*", fields[flight_idx+1]):
            inc_flight_str=fields[flight_idx+1]
            if inc_flight_str.isdigit():
                inc_flight = int(inc_flight_str)
            else:
                if (inc_flight_str[-1] == "," \
                   or inc_flight_str[-1] == ":") and \
                inc_flight_str[:-1].isdigit():
                    inc_flight = int(inc_flight_str[:-1])
    this_incident_df = pd.DataFrame([[dtOfInc,inc_airline, inc_flight]], \
                                   columns=['Date', 'Airline', 'FlightNum'])
    incidents_df = pd.concat([incidents_df,this_incident_df], ignore_index=True)
```

```python
# Drop any rows with missing data
incidents_df.dropna(axis=0, how="any", \
                 inplace=True)
```

```python
# TODO: change column type of 'Date' to datetype

incidents_df['Date'] = pd.to_datetime(incidents_df['Date'])
# change column type of 'FlightNum' to int
incidents_df['FlightNum'] = incidents_df['FlightNum'].astype(int)
# change column type of 'Airline' to str
incidents_df['Airline'] = incidents_df['Airline'].astype(str)

# drop any duplicates from incidents_df
incidents_df.drop_duplicates(inplace=True)
```
